```python
import pandas as pd
import numpy as np
import seaborn as sns
```

### Pandas

Pandas最基本的数据结构是Series和DataFrame,Series类似于一维数组对象,相当于numpy一维数组;而DataFrame是一种二维表格形的数据结构


- 数据结构Series和DataFrame,创建,读取,修改,删除,索引和切片操作,处理缺失值和重复值等清洗操作
- 数据操作函数:数据筛选 排序 合并 聚合 透视
- 数据可视化:Seaborn Plotly
- 时间序列:用于处理时间相关的数据,能够轻松对时间索引的数据进行排序清洗 切片 聚合和频率转换等操作

### 创建

```python
# 列表
df=pd.DataFrame(data=[[1,2,3],[1,2,3]])

df1=df.set_axis(['a','b','c'],axis=1)
df1.set_axis(['A','B'],axis='index')

df2=pd.DataFrame(data=[[1,2,3],[1,2,3]],index=['A','B'],columns=['a','b','c'])
df1,df2
```

```python
# 字典
dict_eg = {'Positive integer': [1, 2, 3, 4, 5], 
           'Greek letter': ['alpha', 'beta', 'gamma', 'delta', 'epsilon']}


df=pd.DataFrame(data=dict_eg)

df.set_index('Greek letter') # 指定索引
```

```python
columns=['X_1','X_2','X_3','X_4']

data=np.random.standard_normal((10,4))

df=pd.DataFrame(data=data,columns=columns)

df
# iloc vs loc iloc用数字索引,loc用名称
df.iloc[:,[1,2]]
df.loc[:,['X_2','X_3']]
```

### 修改数据帧

```python
# rename 对DataFrame的索引标签,列标签进行重命名
# 该方法返回的是一个新的DataFrame,如果在原基础上修改需制定inplace参数
df.rename(columns={'a':'x1','b':'x2','c':'x3'},inplace=True)
df
df.rename(lambda x:f'idx_{x}')
df.rename(lambda x:f'{x}_idx',axis=1)
```

```python
# add_suffix加后缀
# add_prefix加前缀

df.add_prefix('<').add_suffix('>')
```

```python
# reindex 修改列标签顺序
new_col=['x3','x2','x1']

df.reindex(columns=new_col)

df.loc[:,new_col]

df.iloc[:,[2,1,0]]

df.set_axis(new_col,axis=1)
```

```python
# drop删除制定行列
df.drop(columns='x1'),df.drop(index=[0,1])
```

### 四则运算 - 各个列之间

```python

iris_c=iris.rename(columns={'sepal_length':'x1','sepal_width':'x2'})[['x1','x2','species']]
# mean 中
iris_c['x1-E(x1)']=iris_c['x1']-iris_c['x1'].mean()
iris_c['x2-E(x2)']=iris_c['x2']-iris_c['x2'].mean()
iris_c['x1+x2']=iris_c['x1']+iris_c['x2']
iris_c['x1-x2']=iris_c['x1']-iris_c['x2']

iris_c=iris_c.drop(columns=['x1','x2'])
sns.pairplot(iris_c,hue='species',corner=True) # corner只显示对角
```
