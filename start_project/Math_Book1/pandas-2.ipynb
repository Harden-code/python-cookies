## 索引和切片

```python
import pandas as pd
import numpy as np
import seaborn as sns

iris=sns.load_dataset('iris').drop(columns='species')
iris=pd.DataFrame(data=iris.to_numpy(),columns=['x1','x2','x3','x4'])
iris
```

### 索引

```python
# loc[] 选定行列标签来进行索引
# iloc[] 整数索引
iris['x1']
iris.x1
iris.loc[:,'x1']
iris.iloc[:,0]

iris[['x1','x2']]
iris.loc[:,['x1','x2']]
```

### 提取特定列

```python
iris['x1']
iris.x1
```

### 提取特定元素

```python
# iloc[row,col] 
# at[行标签,列标签] iat[行索引,列索引] 用于访问单个元素
# 与loc和iloc不同的是,它返回的是多个元素
iris.at[2,'x1']
iris.iat[2,0]
iris.loc[2,'x1']
```

### 条件索引

条件索引通过布尔条件筛选数据帧里的数据,比如:(df>6) | (df<1.5)

```python
# bool 条件
iris[iris>2].dropna()

# 示例
df=sns.load_dataset('iris')

# 'setosa', 'versicolor', 'virginica'
# 'sepal_length', 'sepal_width', 'petal_length', 'petal_width'
# df['species'].unique()
# df.columns

# 查找类型为setosa sepal_length>2
df[(df['species']=='setosa') & (df['petal_width']>0.2)] 


# 鸢尾花种类 'species' 不在列表 ['virginica','setosa'] 之中
df[~df.species.isin(['virginica','setosa'])] # isin不在列表之中

```

### query 
query(expression)允许通过一定的查询条件来筛选数据,并返回满足的行.

expression:包含bool运算,not,or,and,in 等逻辑运算符号

```python
df.query('species in ["virginica","setosa"]')
df.query('sepal_length > 2.5 and sepal_length<5')
```

### 多层索引
mutil-index 在高维度数据集上进行分层操作和查询

```python
mutil_index=[['A','A','B','B'],range(0,4)]

nums=np.random.standard_normal((4,2))
# 创建多集索引
idx=pd.MultiIndex.from_arrays(mutil_index,names=['c','n'])
data=pd.DataFrame(data=nums,index=idx)

data

data.loc[[('A',0)]] # 索引 A行下的0行
```

```python
data
```

```python
# from_produce 将两个数组生成笛卡尔积
types=['X1','X2']
categories=['A','B','C','D']

# 注意[]里的顺序
idx=pd.MultiIndex.from_product([types,categories],names=['idx1','idx2'])
nums=np.random.standard_normal((8,2))
df=pd.DataFrame(data=nums,index=idx)
df
# 设置索引名称
df.index.set_names('D1',level=0,inplace=True)
df.index.set_names('D2',level=1,inplace=True)
df

# 获取一级索引标签值
df.index.get_level_values(0)

# 获取二级索引标签值
df.index.get_level_values(1)

# 使用xs来对索引进行切片
df.xs('X1',level=0) # 获取D1等于X的所有行
df.xs('A',level=1) # 获取D2等于A的所有行

df.xs(('X1','A'),level=(0,1))
```

```python
# 蒙特卡洛模拟 
date_range=pd.date_range('2024-01-01',  periods=365*2,freq='D')

data=pd.DataFrame(index=date_range)
# 设置随机分子
np.random.seed(0)

num=50

for i in range(num):
    # 生成的随机分布小数
    step_idx=np.random.normal(loc=0.0,scale=1.0,size=len(date_range))
   
    step_idx=np.append(0,step_idx[:-1])
    
    cumsum_idx=step_idx.cumsum() # 累加

    data[f'walk_{i}']=cumsum_idx

data

# data[['walk_0','walk_1']].plot()
data.plot(legend=False,ylim=(-80,80))

    
```
