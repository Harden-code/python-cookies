## 二叉树插入
如果插入数据比当前节点小，并且当前节点的左子节点为空，就将数据插入到左子节点的位置，如果左子树不为空，在递归遍历插入左子树直到找到插入位置

如果插入的数据比当前节点大，并且当前节点右子树为空，就将数据插入到右子节点位置，如果不为空在递归遍历右子树，直到找到插入位置

## 二叉树删除
要删除节点没子节点，只需要将父节点中指向要删除节点的指针为null

要删除节点只有一个节点，主需要更新父节点中要删除节点的指针，让它重新指向要删除节点即可

要删除节点有两个节点，需要找到这个节点的右子树最小节点，把它替换到要删除的节点上（或左子树最大节点，总之就是最接近这个删除节点的节点）。然后删除这个最小节点，因为最小节点没有左子树（可能有右子节），所以可以根据上面两条规则来删除最小节点


````python
from typing import Optional
class Node:
    def __init__(self,
                 val:int,
                 left:Optional['Node']=None,
                 right:Optional['Node']=None) -> None:
        self.val=val
        self.left=left
        self.right=right

def search(root:Node,val:int)->Node:
    if not root: 
        return None
    if val>root.val:
        return search(root.right,val)
    elif val <root.val:
        return search(root.left,val)
    else:
        return root

def insert(root:Node,node:Node):
    if node.val<root.val :
        if not root.left:
            root.left=node
        else:
            insert(root.left,node)
    else:
        if not root.right:
            root.right=node
        else:
            insert(root.right,node)

def delete(root:Node,val:int):
    '''
        先找到要删除的节点
        1.要删除的节点有两个字节点，找到右子树最左子节点替换
        2.删除只有一个节点或者没有子节点的
    '''
    p=root
    pp=root # 记录删除节点的前一个节点
    while p and p.val!=val:
        pp=p
        if val > p.val:
            p=p.right
        else:
            p=p.left
    if p==None: # 未找到节点不删除
        return
    if p.left and p.right:
        t=p.right
        tt=p
        while t.left :
            tt=p
            p=p.left
        p.val=t.val
        p=t
        pp=tt

    chirld=None
    if p.left:
        chirld=p.left
    elif p.right:
        chirld=p.right

    # 判断删除节点是在父节点那边
    if not pp: # 根节点
        root=chirld 
    elif pp.left.val==val:
        pp.left=chirld
    else :#pp.right.val==val
        pp.right=chirld

node2=Node(1)
node1=Node(2,node2)
node3=Node(4)
root=Node(3,node1,node3)

print(search(root,1).val)
node4=Node(5)
insert(root,node4)

import pdb
pdb.set_trace()
print(root)
````
