````python
from llama_index.core import SimpleDirectoryReader,DocumentSummaryIndex
from llama_index.llms.openai import OpenAI
from llama_index.core.node_parser import LangchainNodeParser,SimpleNodeParser
from langchain.text_splitter import SpacyTextSplitter

splitter=SpacyTextSplitter(pipeline='zh_core_web_sm',chunk_size=2048)
parse=LangchainNodeParser(lc_splitter=splitter)

doc=SimpleDirectoryReader('./md/').load_data()
# parse.get_nodes_from_documents(doc)
chatgpt = OpenAI(temperature=0, model="gpt-3.5-turbo")
index=DocumentSummaryIndex.from_documents(doc,llm=chatgpt,
                                    transformations=[parse],
                                    show_progress=True)
````

````python
index.get_document_summary('3fb86b69-5543-4052-aa0d-c89f9dfd7f99')
````

````python
engine=index.as_query_engine(llm=chatgpt)
engine.query('用中文总结这个故事')
````

````python
# https://docs.llamaindex.ai/en/stable/examples/vector_stores/FaissIndexDemo/
import faiss
from langchain.embeddings.huggingface import HuggingFaceEmbeddings
from langchain.text_splitter import SpacyTextSplitter
from llama_index.core import ServiceContext,SimpleDirectoryReader,VectorStoreIndex,StorageContext
from llama_index.core.node_parser import LangchainNodeParser
from llama_index.embeddings.langchain import LangchainEmbedding
from llama_index.vector_stores.faiss import FaissVectorStore

doc=SimpleDirectoryReader('./data/').load_data()

splitter=SpacyTextSplitter(pipeline='zh_core_web_sm',chunk_size=1000)
dimension=768
faiss_index=faiss.IndexFlatIP(dimension)

em_model=LangchainEmbedding(langchain_embeddings=HuggingFaceEmbeddings(model_name='sentence-transformers/paraphrase-multilingual-mpnet-base-v2',device='cpu'))

context=ServiceContext.from_defaults(embed_model=em_model)


vector=FaissVectorStore(faiss_index=faiss_index)

storage_context=StorageContext.from_defaults(vector_store=vector)

index=VectorStoreIndex.from_documents(documents=doc
                                      ,storage_context=storage_context
                                      ,service_context=context)

````

````python
import spacy
'''
sentence segmentation 
通过sents来查看子句.通过has_annotationdd的SENT_START
来检查是否含有子句
'''

nlp=spacy.load('zh_core_web_sm')
doc=nlp('''有不少人在使用OpenAI提供的GPT系列模型的时候，都反馈效果并不好。这些反馈中有一大类问题，是回答不了一些简单的问题。比如当我们用中文问 AI 一些事实性的问题，AI 很容易胡编乱造。而当你问它最近发生的新闻事件的时候，它就干脆告诉你它不知道 21 年之后的事情。''')

assert doc.has_annotation('SENT_START')

for segment in doc.sents:
    print(segment.text)
````
