```python
# 使用deque限制元素个数
import os
from collections import deque
d = deque(maxlen=3)
d.append(1)
a = [2, 3, 4]
d.extend(a)  # [2,3,4]

# 搜索匹配的元素,并保留最后的搜索记录


def search(lines: list[str], patten: list[str], history: deque):
    for idx, line in enumerate(lines):
        if line in patten:
            yield idx, line
        history.append(line)


dirs = os.listdir('/')
patten = ['home', 'cores']
history = deque(maxlen=3)
for idx, line in search(dirs, patten, history):
    print(idx, line)

for i in history:
    print(i)
```

```python
# 优先级队列
from typing import TypeVar
import heapq
I = TypeVar('I')


class PriorityQueue:
    def __init__(self) -> None:
        self._queue = []
        self._idx = 0

    def push(self, item: I, val: int = 1):
        idx = self._idx
        self._idx += 1
        # 使用元组 可以进行排序
        heapq.heappush(self._queue, (val, idx, item))

    def pop(self):
        return heapq.heappop(self._queue)[-1]

    def __repr__(self) -> str:
        return f'{self._queue}'


queue = PriorityQueue()

queue.push('a', 1)
queue.push('b', 2)
queue.push('c', 3)

print(queue, queue.pop())
```
