```python
# 4.x+y+3z=6 x-y+z=4 ;z=2 z=0 half between?
A=sympy.Matrix([[1,1,3],[1,-1,1],[0,0,1]])
b=sympy.Matrix([6,4,0])
sympy.pprint(A.solve(b))

b=sympy.Matrix([6,4,2])
sympy.pprint(A.solve(b))

a=np.linspace(-5,5,10)
X,Y=np.meshgrid(a,a)


# 创建一个 3D 绘图对象
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# -x/3 - y/3 + 2
expr1=sympy.solve(x+y+3*z-6,z)

# -x + y + 4
expr2=sympy.solve(x-y+z-4,z)

Z1=-X/3-Y/3+2

Z2=-X+Y+4

ax.plot_surface(X,Y,Z1,alpha=0.5,color='r')
ax.plot_surface(X,Y,Z2,alpha=0.5,color='b')


# x+y+3z=6 画相交线解方程组
# x-y+z=4
ys=np.linspace(-5,5,100)
xs, zs = [], []
A1=np.array([[1,1,3],[1,-1,1]])
b=np.array([6,4])
for y_val in ys:
   # 把y移到右边 
   B = b - y_val * np.array([1, -1])
   solution = np.linalg.solve(A1[:, [0, 2]], B) # ⚠️ 系数
   xs.append(solution[0])
   zs.append(solution[1])

ax.plot(xs, ys, zs, color='green', label='Intersection line')

ax.scatter(5,1,0,s=2)
ax.scatter(1,-1,2,s=2)
# ax.set_aspect('equal')
```
