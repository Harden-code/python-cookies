```python
# 等距线
from matplotlib.patches import Circle
import matplotlib.pyplot as plt
rad=np.linspace(1,5,5)
fig,ax=plt.subplots()

# 创建颜色列表，每个圆一个颜色
colors = ['blue', 'green', 'red', 'purple', 'cyan']

# 在相同中心添加多个圆
for r, color in zip(rad, colors):
    c = Circle(xy=(0, 0), radius=r, facecolor='none', edgecolor=color, linewidth=2)
    ax.add_patch(c)

ax.set_xlim(-6, 6)
ax.set_ylim(-6, 6)
ax.set_aspect('equal')
```

```python
a=np.linspace(-5,5,1001)
X,Y=np.meshgrid(a,a)

height=X**2+Y**2

plt.contour(X,Y,height,levels = np.arange(11), cmap='RdYlBu_r')

ax=plt.gca()

ax.axhline(y=0, color='k', linewidth = 0.25)
ax.axvline(x=0, color='k', linewidth = 0.25)
ax.set_xlim(-12, 12)
ax.set_ylim(-12, 12)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

ax.set_aspect('equal',adjustable='box')
```

## 向量积

### 术语

vector product 向量积

cross product 叉积

> 向量积结果为向量

$a=a_1i+a_2j+a_3k$

$b=b_1i+b_2j+b_3k$

$a*b=a_1b_1(i*i)+a_1b_2(i*xj)+a_1b_3(j*k)+a_2b_1(j*i)+a_2b_2(j*j)+a_2b_3(j*k)+a_3b_1(k*i)+a_3b_2(k*j)+a_3b_3(k*k)$

=$(a_2b_3-a_3b_2)i+(a_3b_1-a_1b_3)j+(a_1b_2-a_2b_1)k$

> numpy.cross()

```python
a=np.array([1,2])
b=np.array([2,2])
np.cross(a,b)
```
