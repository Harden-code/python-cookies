### 艺术家
在matplotlib中,艺术家(artist)指图形的每个可见元素,如图脊,图轴,坐标系,标题,图例,线条,网格,色块等等,每个艺术家对象都有自己的默认属性和方法,控制其外观和行为

### 图脊
图脊(spine)指的是图形中边框线,用于界定图形的边界,由四条边框组成:上脊(top spine),下脊(bottom spine),左脊(left spine),右脊(right spine)

### 图轴
图轴(axis)指的是图形中的坐标轴,用于表示数据数值范围和刻度.图轴包含轴线(axis line),刻度(tick),轴标签(axis label)等艺术家

### 注释
- 标签(title):matplotlib.pyplot.title()或ax.set_title()
- 横轴标题(x axis label):matplotlib.pyplot.xlabel() 或 ax.set_xlabel()
- 纵轴标题(y axis label):matplotlib.pyplot.ylabel() 或 ax.set_ylabel()
- 图例(legend):matplotlib.pyplot.
- 文字(text):matplotlib.pyplot.text()
- 注释(annotation):matplotlib.pyplot.annotate()

### 视觉
ax.view_init(elev,azim,roll),用于设置三维坐标轴的视觉,也叫照相视觉,由是三个参数

- 仰角(elevation):定义观察者与xy平面的夹角,当elevation为正向上倾斜,负数向下倾斜
- 方位角(azimuth):定义观察者绕z轴旋转的角度,它决定了观察者xy平面上的位置
- 滚动角(roll):定义了观察者视线方向绕旋转的角度,正值向右倾斜,负值向左倾斜

设置正交投影
ax.set_proj_type('orthi') 

### 风格
图像风格指的是图标的整体外观和样式,包括配色方案,线条类型,图脊,图轴,线条宽度,字体,标记符号等

- default:蓝色线条和绿色网格
- classic:使用黑色线条和白色背景
- ggplot:使用灰色网格和彩色线条
- fivethirtyeight:使用红色和蓝色线条,以及灰色网格
- dark_background:深色背景和亮色线条,黑暗主体环境
- seaborn:模仿seaborn,使用柔和颜色和灰色网格

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator,MultipleLocator,FormatStrFormatter
import numpy as np
```

### 修改图脊

```python
x_arr=np.linspace(-4,4,200)
y_arr=x_arr*np.exp(-x_arr**2)
```

```python
fig=plt.figure(figsize=(6,6),tight_layout=True)

ax1=fig.add_subplot(3,2,1)
ax1.plot(x_arr,y_arr)
ax1.set_xticks([])
ax1.set_yticks([])


ax2=fig.add_subplot(3,2,2)
ax2.plot(x_arr,y_arr)
ax2.set_xticks([])
ax2.set_yticks([])
ax2.spines[['top','bottom']].set_visible(False)
ax2.spines['left'].set_color(None)

ax3=fig.add_subplot(3,2,3)
ax3.plot(x_arr,y_arr)
# ax3.set_xticks([])
# ax3.set_yticks([])
ax3.spines['top'].set_visible(False)
ax3.yaxis.set_ticks_position('right')
ax3.xaxis.set_ticks_position('top')

ax4=fig.add_subplot(3,2,4)
ax4.plot(x_arr,y_arr)
ax4.set_xticks([])
ax4.set_yticks([])
# outward,axes[0,1]之间,data 注意是元组
ax4.spines['bottom'].set_position(('data',0))
ax4.spines['left'].set_position(('data',0))
ax4.spines[['top','right']].set_color('none')

ax5=fig.add_subplot(3,2,5)
ax5.plot(x_arr,y_arr)
ax5.spines['bottom'].set_position(('axes',0.5))

ax5.set_xticks([])
ax5.set_yticks([])


ax6=fig.add_subplot(3,2,6)
ax6.plot(x_arr,y_arr)
ax6.set_xticks([])
ax6.set_yticks([])
ax6.spines['bottom'].set_position(('outward',10))
ax6.spines['bottom'].set(edgecolor='r',lw=4)
```

### 修改网格

```python

fig=plt.figure(figsize=(6,6),tight_layout=True)

ax1=fig.add_subplot(3,2,1)
ax1.plot(x_arr,y_arr)
# 打开网格
ax1.grid(True)

ax2=fig.add_subplot(3,2,2)
ax2.plot(x_arr,y_arr)
ax2.grid(linestyle='-',lw=0.5,color='red')

ax3=fig.add_subplot(3,2,3)
ax3.plot(x_arr,y_arr)
ax3.set_axisbelow(True)
ax3.minorticks_on()
ax3.grid(which='major',linestyle='-',
         lw=0.5,color='r')
ax3.grid(which='minor',linestyle=':',
         lw=0.5,color='black')

ax4=fig.add_subplot(3,2,4)
ax4.plot(x_arr,y_arr)
ax4.grid(which='major',color='r',
         lw=0.5,linestyle='-')
ax4.grid(which='minor',color='black',
         lw=0.5,linestyle=':')
ax4.minorticks_on()
ax4.tick_params(which='both',top='off'
                ,left='off'
                ,right='off'
                ,bottom='off')

ax5=fig.add_subplot(3,2,5)
ax5.plot(x_arr,y_arr)

for tick in ax5.yaxis.get_major_ticks():
    tick.tick1line.set_visible(False)
    tick.tick2line.set_visible(False)
    tick.label1.set_visible(False)
    tick.label2.set_visible(False)

ax6=fig.add_subplot(3,2,6)
ax6.plot(x_arr,y_arr)
for tick in ax6.xaxis.get_major_ticks():
    tick.tick1line.set_visible(False)
    tick.tick2line.set_visible(False)
    tick.label1.set_visible(False)
    tick.label2.set_visible(False)
```

### 对数坐标系

```python
X=np.linspace(0.001,90,5000)
```

```python
fig=plt.figure(figsize=(8,8))
# 定位器
n2_locator=MultipleLocator(2.0)
n4_locator=MultipleLocator(0.4)

ax1=fig.add_subplot(2,2,1,xlim=(0.,10),ylim=(0.,10))
ax1.plot(X,10**X,color='C0')
ax1.plot(X,X,color='C1')
ax1.plot(X,np.log10(X),color='C2')

ax1.xaxis.set_major_locator(n2_locator)
ax1.xaxis.set_minor_locator(n4_locator)
ax1.yaxis.set_major_locator(n2_locator)
ax1.yaxis.set_minor_locator(n4_locator)
ax1.grid(True,'major',color='0.85',lw=0.5,zorder=-20)
ax1.grid(True,'minor',color='0.65',lw=0.75,zorder=-10)
ax1.tick_params(axis='both',labelbottom=False,bottom=False)

ax1.text(1.25,8.50,'$f(x)=10^x$',color='C0',fontname='roboto')
ax1.text(5.25,5.00,'$f(x)=x$',color='C1',fontname='roboto')
ax1.text(5.50,1.50,'$f(x)=log_{10}(x)$',color='C2',fontname='Roboto')

ax2=fig.add_subplot(2,2,2,xlim=(0.001,100),ylim=(0.0,10),sharey=ax1)
ax2.set_xscale('log')
ax2.tick_params('both',labelbottom=False,bottom=False)
ax2.tick_params('both',labelleft=False,left=False)
ax2.plot(X, 10 ** X, color="C0")
ax2.plot(X, X, color="C1")
ax2.plot(X, np.log10(X), color="C2")
ax2.grid(True,'major',color='0.85',lw=0.5,zorder=-20)
ax2.grid(True,'minor',color='0.65',lw=0.75,zorder=-10)

ax3=fig.add_subplot(2,2,3,xlim=(0.0,10),ylim=(0.001,100),sharex=ax1)
ax3.set_yscale('log')
ax3.plot(X,10**X,color='C0')
ax3.plot(X,X,color='C1')
ax3.plot(X,np.log10(X),color='C2')
ax3.grid(True,'minor',color='0.85',lw=0.5,zorder=-20)
ax3.grid(True,'major',color='0.65',lw=0.75,zorder=-10)

ax4=fig.add_subplot(2,2,4,xlim=(0.001,100),ylim=(0.001,100),sharex=ax2,sharey=ax3)
ax4.set_xscale('log')
ax4.set_yscale('log')
ax4.tick_params(which='both',labelleft=False,left=False)
ax4.plot(X, 10 ** X, color="C0")
ax4.plot(X, X, color="C1")
ax4.plot(X, np.log10(X), color="C2")
ax4.grid(True, "minor", color="0.85", linewidth=0.50, zorder=-20)
ax4.grid(True, "major", color="0.65", linewidth=0.75, zorder=-10)
```

```python
x_array = np.linspace(-4, 4, 200)
y_array = x_array*np.exp(-x_array**2)
y_array_2 = np.exp(-x_array**2)/2
```

```python
fig=plt.figure(figsize=(8,12),tight_layout=True)

ax=fig.add_subplot(3,2,1)
ax.plot(x_array,y_array)
ax.set_xlabel('x axis label')
ax.set_ylabel('y axis label')
ax.set_xlim(-4,4)
ax.set_ylim(-0.5,0.5)
ax.grid(True)

ax=fig.add_subplot(3,2,2)
ax.plot(x_array,y_array)
ax.set_xlabel('x axis label',loc='left')
ax.set_ylabel('y axis label',loc='top')
ax.set_xlim(-4,4)
ax.set_ylim(-0.5,0.5)
ax.grid(True)

ax=fig.add_subplot(3,2,3)
ax.plot(x_array,y_array)
ax.set_xlabel('x axis label',loc='right')
ax.set_ylabel('y axis label',loc='bottom')
ax.set_xlim(-4,4)
ax.set_ylim(-0.5,0.5)
ax.grid(True)

ax=fig.add_subplot(3,2,4)
ax.plot(x_array,y_array)
ax.xaxis.set_label_position('top')
ax.xaxis.tick_top()
ax.set_xlabel('x axis label')
ax.set_ylabel('y axis label')
ax.set_xlim(-4,4)
ax.set_ylim(-0.5,0.5)
ax.grid(True)

ax=fig.add_subplot(3,2,5)
ax.plot(x_array,y_array)
ax.set_xlabel('x axis label')
ax.set_ylabel('y axis label')
ax.set_title('this is the tile',loc='left')
ax.set_xlim(-4,4)
ax.set_ylim(-0.5,0.5)
ax.grid(True)

ax=fig.add_subplot(3,2,6)
ax.plot(x_array,y_array,label='f1')
ax.plot(x_array,y_array_2,label='f2')
ax.set_xlim(-4,4)
ax.set_ylim(-0.5,0.5)
ax.legend()
ax.grid(True)
ax.set_title('This is the tile \n In case multiple lines', loc = 'left')
ax.set_xlabel('x axis label')
ax.set_ylabel('y axis label')
```

### 三维网格

```python
from mpl_toolkits.mplot3d import axes3d
X,Y,Z=axes3d.get_test_data(0.05)
```

```python
fig=plt.figure(figsize=(12,12),constrained_layout=True)
angle_arr=np.linspace(0,180,13)
num_grid=len(angle_arr)
gspec=fig.add_gridspec(num_grid,num_grid)
# 获取行和列
nrow,ncol=gspec.get_geometry()

axes=np.array([[fig.add_subplot(gspec[i,j],projection='3d')for j in range(ncol)] for i in range(nrow) ])

for i in range(nrow):
    elev=angle_arr[i]
    for j in range(ncol):
        azim=angle_arr[j]

        axes[i,j].plot_wireframe(X,Y,Z,rstride=10,cstride=10)
        axes[i,j].set_proj_type('ortho')
        axes[i,j].grid('off')
        axes[i,j].set_xticks([])
        axes[i,j].set_yticks([])
        axes[i,j].set_zticks([])
        axes[i,j].set_xlim(X.min(),X.max())
        axes[i,j].set_ylim(Y.min(),Y.max())
        axes[i,j].set_zlim(Z.min(),Z.max())

        axes[i,j].view_init(elev=elev,azim=azim)
```
