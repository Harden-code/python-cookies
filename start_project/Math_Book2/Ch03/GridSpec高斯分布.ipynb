```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
mean=[0,0]
cov=[[1,1],[1,2]]
x,y=np.random.multivariate_normal(mean,cov,1000).T
```

```python
fig=plt.figure(figsize=(6,6))
grid=plt.GridSpec(4,4,hspace=0.3,wspace=0.3)
main_ax=fig.add_subplot(grid[:-1,1:])
y_hist=fig.add_subplot(grid[:-1,0],xticklabels=[])
x_hist=fig.add_subplot(grid[-1,1:],yticklabels=[])

main_ax.plot(x,y,'ob',markersize=3,alpha=0.3)
main_ax.set_xticks([])
main_ax.set_yticks([])

x_hist.hist(x,40,orientation='vertical',color='b'
            ,edgecolor='white',lw=0.5)
x_hist.invert_yaxis()
x_hist.set_xticks([])
x_hist.set_yticks([])

y_hist.hist(y,40,orientation='horizontal',color='b'
            ,edgecolor='white',lw=0.5)

y_hist.set_xticks([])
y_hist.set_yticks([])

y_hist.invert_xaxis()
```
