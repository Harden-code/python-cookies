```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_arr=np.linspace(-5,5,200)
mu_arr=[-2,-1,0,1,2]
sigma_arr=[0.5,0.75,1,1.5,2]

mu_arr_,sigma_arr_=np.meshgrid(mu_arr,sigma_arr)
```

```python
fig,ax=plt.subplots(ncols=5,nrows=5,figsize=(10,10))

for mu,sigma,ax in zip(mu_arr_.ravel(),sigma_arr_.ravel(),ax.ravel()):
    title=f'mu={str(mu)} sigma={str(sigma)}'
    ax.plot(x_arr,norm.pdf(x_arr,mu,sigma))
    ax.fill_between(x_arr,norm.pdf(x_arr,mu,sigma),0,color='#DBEEF3')

    ax.set_yticks([])
    ax.set_yticklabels([])

    ax.get_yaxis().set_visible(False)
    ax.set(ylabel=None)

    ax.set_title(title)
    ax.set_xlim(-5,5)
    ax.set_ylim(0,1)

    ax.set_xticks([-5,0,5])
    ax.spines['left'].set_color('none')
    ax.spines['right'].set_color('none')
    ax.spines.top.set_visible(False)

    ax.xaxis.set_ticks_position('bottom')
    ax.tick_params(axis='x',direction='in')
# 
plt.tight_layout()
```
