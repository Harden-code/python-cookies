```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

data = load_iris()

iris = pd.DataFrame(data.data, columns=['SpealLength', 'Spealwidth', 'PetalLength', 'PetalLength'])
print(iris.head())
X = np.c_[iris['SpealLength'], iris['Spealwidth']]
kmeans = KMeans(n_clusters=3)
#
cluster = kmeans.fit(X)
label = cluster.labels_
print("label", label)
print('centers', cluster.cluster_centers_)

colors = ['r.', 'g.', 'b.']
fig, ax = plt.subplots()
for i in range(len(data.data)):
    ax.plot(iris['SpealLength'][i], iris['Spealwidth'][i], colors[label[i]])

plt.scatter(cluster.cluster_centers_[:, 0], cluster.cluster_centers_[:, 1], marker='x', s=150, linewidths=5, zorder=10,
            color='k')

plt.show()
```
