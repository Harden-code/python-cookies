```python
import matplotlib.pyplot as plt
import numpy as np

def draw_quiver(xy,ax,origin=(0,0),c='black'):
    ax.quiver(*origin,*xy,angles='xy',scale_units='xy',scale=1,color=c)


fig,ax = plt.subplots()
a=np.array([1,4])
b=np.array([3,1])
draw_quiver(a,ax)
draw_quiver(b,ax)
c=(a+b)/2
draw_quiver(c,ax)

d=3/4*a+1/4*b
draw_quiver(d,ax)

f=1/4*a+1/4*b
draw_quiver(f,ax)

g=-a+2*d
draw_quiver(g,ax,c='red')

ax.plot([1,3],[4,1])
ax.set_xbound(0,5)
ax.set_ybound(0,5)
ax.set_aspect('equal',adjustable='box')

# c+d=1 都会在a->b 方向的向量上
plt.show()
```
