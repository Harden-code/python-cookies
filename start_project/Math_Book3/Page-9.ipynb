### 离心率
不同的圆锥通过离心率e联系起来
> $x_2^2 = 2px_1+ (e^2-1)x_1^2$\

圆锥曲线上的一点到平面内一定点的距离与到不过此点的一定直线的距离之比。其中此定点称为焦点，而此定直线称为准线。

圆锥曲线离心率与轴长关系
> $e=\frac{c}{a}$;其中c为半焦距,a为半长轴(椭圆)或半实轴(双曲线)

- 圆: `e=0`
- 椭圆: `0<e<1`
- 抛物线: `e=1`
- 双曲线: $1<e< \infty$

```python

import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as patches

x=np.linspace(-4,4,num=201)
y=np.linspace(-4,4,num=201)

m=1
n=1.5

xx,yy=np.meshgrid(x,y)

e_array=np.linspace(0,3,num=31)

colors=plt.cm.RdYlBu(np.linspace(0,1,len(e_array)))
fig,ax=plt.subplots(figsize=(8,8))
print(type(colors))
for i in range(0,len(e_array)):
    e=e_array[i]

    ellipse=yy**2-(e**2-1)*xx**2-2*xx

    color_code=colors[i,:].tolist()

    plt.contour(xx,yy,ellipse,levels=[0],colors=[color_code])

# 两条垂直线
plt.axhline(y=0,color='k',linestyle='-')
plt.axvline(x=0,color='k',linestyle='-')

ax.set_xticks([])
ax.set_yticks([])

ax.set_xlim([-4,4])
ax.set_ylim([-4,4])

ax.spines['top'].set_visible=False
ax.spines['right'].set_visible=False
ax.spines['bottom'].set_visible=False
ax.spines['left'].set_visible=False

plt.show()
```

### 公式变化
> ${\frac{x_1^2}{m^2}}$ + $\frac{x_2^2}{n^2}$ - $2p$ $\frac{x1x2}{mn}$  = 1 

上述可以看作是椭圆和双曲线的叠加,$x_1 x_2=1$ 实际上就是一个旋转的双曲线,参数p可以看成调节双曲线影响力的参数,p越大双曲线的影响力越强

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(-4,4,num=201)
y=np.linspace(-4,4,num=201)

xx,yy=np.meshgrid(x,y)

#a=2
#b=1
#p_array=np.linspace(0,2,num=21)

a=1
b=1
p_array=np.linspace(0,-2,num=21)

fig,ax=plt.subplots(figsize=(8,8))

colors=plt.cm.RdYlBu(np.linspace(0,1,len(p_array)))

for i in range(0,len(p_array)):
    p=p_array[i]
    ellipse=(xx/a)**2 + (yy/b)**2 - 2*p*(xx/a)*(yy/b)
    color_code=colors[i,:].tolist()
    plt.contour(xx,yy,ellipse,levels=[1],colors=[color_code])

ax.axvline(x=0,color='k',linestyle='-')
ax.axhline(y=0,color='k',linestyle='-')
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlim([-4,4])
ax.set_ylim([-4,4])
ax.spines['top'].set_visible=False
ax.spines['bottom'].set_visible=False
ax.spines['left'].set_visible=False
ax.spines['right'].set_visible=False
```

### 特殊椭圆:和给给定矩形相切
> $\frac{x_1^2}{m^2} + \frac{x_2^2}{n^2} - \frac{2px_1x_2}{mn} = 1 - p^2$

转换成
> $\frac{1-p^2}{1} ( \frac{x_1^2}{m^2} + \frac{x_2^2}{n^2} - \frac{2px_1x_2}{mn} )$ = 1 

1. 当p等于0时 椭圆为正圆
2. 当p等于$\pm 1$时 退化成一条直线
3. 当m=n时 椭圆形状随着参数p变化,当p越靠近0椭圆越接近正圆;p的绝对值越靠近1椭圆形状越扁


```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as patches

x=np.linspace(-4,4,num=201)
y=np.linspace(-4,4,num=201)

m=1
n=1.5
xx,yy=np.meshgrid(x,y)

rho_array=np.linspace(-0.95,0,num=20)

fig,ax=plt.subplots(figsize=(8,8))

# facecolor 面部颜色,消除填充色
rect=patches.Rectangle((-m,-n),2*m,2*n,linewidth=0.25,edgecolor='k',linestyle='-',facecolor='none')

colors=plt.cm.RdYlBu(np.linspace(0,1,num=len(rho_array)))
ax.add_patch(rect)

for i in range(0,len(rho_array)):
    rho=rho_array[i]
    color_code=colors[i,:].tolist()
    ellipse=((xx/m)**2 - 2*rho*(xx/m)*(yy/n) + (yy/n)**2)/ (1-rho**2)

    plt.contour(xx,yy,ellipse,levels=[1],colors=[color_code])

plt.axvline(x=0,color='k',linestyle='-')
plt.axhline(y=0,color='k',linestyle='-')

ax.set_xticks([])
ax.set_yticks([])

ax.set_xlim([-2,2])
ax.set_ylim([-2,2])

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)


```

### 超椭圆
超椭圆(superellipse)是对椭圆的扩展
> $\left\vert \frac{x_1}{a} \right\vert ^ p + \left\vert \frac{x_2}{b} \right\vert ^ p = 1$

一般情况下p为大于0数值,当p=2时为椭圆,当p=1时超椭圆为棱形
> $\left\vert \frac{x_1}{a} \right\vert + \left\vert\ \frac{x_2}{b} \right\vert = 1$

### 超椭圆

> ( $\left\vert \frac{x_1}{a} \right\vert + \left\vert\ \frac{x_2}{b} \right\vert ) ^ \frac{t}{r} + \left\vert \frac{x_3}{c} \right\vert ^ t = 1$


```python
import matplotlib.pyplot as plt
import numpy as np 
a=1
b=1

p=[0.5,1,2,3]
q=p

pp,qq=np.meshgrid(p,q)
# flatten 扁平
pp=pp.flatten()
qq=qq.flatten()

x1=np.linspace(-2,2,num=101)
x2=x1

xx1,xx2=np.meshgrid(x1,x2)

fig,axes=plt.subplots(ncols=4,nrows=4,figsize=(12,12))
# axes.flat
for p,q,ax in zip(pp,qq,axes.flat):
    if np.isinf(p):
        zz=np.maximum(np.abs(xx1/a),np.abs(xx2/b))
    else:
        #zz=((np.abs((xx1/a))**p)+(np.abs((xx2/b))**q))**(1./q)
        zz=((np.abs((xx1/a))**p)+(np.abs((xx2/b))**q))
    
    ax.contourf(xx1,xx2,zz,20,cmap='RdYlBu_r')
    ax.contour(xx1,xx2,zz,[1],colors='k',linewidths=2)

    ax.axvline(x=0,color='k',linewidth=0.25)
    ax.axhline(y=0,color='k',linewidth=0.25)
    ax.set_xlim(-2,2)
    ax.set_ylim(-2,2)
    ax.spines['top'].set_visible=False
    ax.spines['bottom'].set_visible=False
    ax.spines['left'].set_visible=False
    ax.spines['right'].set_visible=False
    ax.set_xlabel('$x_1$')
    ax.set_ylabel('$x_2$')
    ax.set_title(f'p={str(p)},q={str(q)}')
    ax.set_aspect('equal',adjustable='box')
```

### 双曲函数
当a=1,b=1时双曲线为单位双曲线,单位双曲线可以用来定义双曲函数

$\sinh a = \frac{\exp t - \exp t}{2}$

$\cosh a = \frac{\exp t + \exp t}{2}$

$\tanh a = \frac{\exp t - \exp t}{\exp t + \exp t}$

### 圆锥曲线一般式
> $Ax_1^2 + Bx_1x_2 + Cx_2^2 + Dx_1 + Ex_2 + F =0 $

当为正圆
> $Ax_1^2 + Cx_2^2 + Dx_1 + Ex_2 + F =0$,A=C 

当为正椭圆
> $Ax_1^2 + Cx_2^2 + Dx_1 + Ex_2 + F =0$,A!=C,AC>0

当为正双曲线
> $Ax_1^2 + Cx_2^2 + Dx_1 + Ex_2 + F =0$,AC<0

当为正抛物线
> $\begin{cases}
    Ax_1^2 + Dx_1 + Ex_2 + F = 0 \\
    Cx_2^2 + Dx_1 + Ex_2 + F = 0 
\end{cases}$
