### 代数函数
通过常数与自变量相互之间有限次的加减乘除有理指数幂和开方构造的函数

### 一次函数:一条斜线
有4种不同方式构造
1. 斜截式:slope-intercpt form
    > $y=f(x)=ax+b$,一次函数有两种斜率正斜率和负斜率,一个大于0单调递增,一个小于0单调递减.斜率越大一次函数图更陡峭
2. 点斜式:point-slope form
    > $y-y_1=f(x)-y_1=a(x-x_1)$,给定斜率a和直线上的点(x,y)便可确定平面上的一条线
3. 两点式:two-point form
    > $y-y_1=\frac{y-y_1}{x-x_1}(x-x_1)$,通过两点确定一条直线
4. 参数方程:parametic equation

### 二次函数:一条抛物线
二次函数(quadratic funcion)是二次多项式函数(second order polynomoal function),二次函数是一套抛物线,可以开口向上或开口向下,对称轴平行于纵轴

三种形式
1. 基本上(standard form)
    > $f(x)=ax^2+bx+c,a!=0$,a为二次项系数,b为一次项系数,c为常数项.二次函数为凹凸函数,顶点位置对应最大(maxima)或最小值(minima).二次函数开口随着系数a的变化,a的绝对值越大,开口越小
2. 两根式(facored form)
    > $f(x)=a(x-r_1)(x-r_2),a!=0$
3. 顶点式(vertex form)
    > $f(x)=a(x-h)^2+k,a!=0$

### 多项式函数:从函数叠加角度
多项式函数(polynomial function)
> $y=f(x)=a_kx^k + a_{k-1}x^{k-1}+a_{k-2}x^{k-2} = \sum_{i=1}^ka_ix^i$

三次函数(cubic function)
> $y=f(x)=a_3x^3+a_2x^2+a_1x+a_0$

### 幂函数:底数为自变量
> $f(x)=kx^p$;

x为底数(base),p为指数(exponent或power).幂函数就是一个数和它自己相乘的积

- 当p为偶数时,幂函数为偶函数图像关于y轴对称,p越大x的绝对值增大,函数值越快接近正无穷

- 当p为奇数时,幂函数为偶函数图像关于圆点对称,p越大x的绝对值增大,函数值越快接近正无穷

### 平方根函数
> $y=f(x)=\sqrt[2]{x}=x^{\frac{1}{2}}$

numpy中用numpy.sqrt(x)来计算平方根,也可用x**(1/2)

### 立方根
> $y=f(x)=\sqrt[3]{x}=x^{\frac{1}{3}}$

numpy中用cbrt(x)用来计算立方根,也可用x**(1/3)

### 反比例函数
> $y=f(x)=\frac{k}{x}$

当$\left\vert k \right\vert>1$时,双曲线朝远离原点方向拉伸;当$\left\vert k \right\vert<1$双曲线向原点方向压缩

反比例函数移动之后可以得到最简单的有理函数
> $f(x)=\frac{k}{x-h}+a$

### 渐进线
渐进线指与曲线极限相关的一条直线,当曲线上某动点沿该曲线的一个分支移向无穷时,动点到该渐进线的垂直距离趋近于0;比如反比例函数两条渐进线,水平渐进线y=0和竖直渐进线x=0

### 分段函数
分段函数是一类不连续函,分段函数指对应自变量x的不同取值范围,有不同的解析函数
> $f(x)=\begin{cases}
        4 & x<-2 \\
        -1 & -2 <= x < 3 \\
        3 & 3 <=x
        \end{cases}$

### 绝度值函数
绝对值函数可以看做分段函数
> $f(x)=k\left\vert x-h \right\vert+a$

numpy.absolute来计算绝对值

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0.1,10,num=1000)

plt.plot(x,np.sqrt(x),color='r')

plt.yscale('log') # 设置刻度尺为log
ax=plt.gca()
ax.set_box_aspect(1)#箱形图的比例设置为1，形成了一个宽高比为1:1的长方形形状。
```
