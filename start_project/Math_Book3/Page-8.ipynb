### 圆锥曲线
圆锥曲线通过一个对顶圆锥和一个截面(cutting plane)相交得到一系列曲线,圆锥曲线主要分为:正圆(circle),椭圆(ellipse),抛物线(parabola),双曲线(hyperbola)、
#### 退化圆锥曲线
退化双曲线(degenerate parabola)为两条相交直线
退化抛物线(degenerate hyperbola)可以为直线或者两条平行线
退化椭圆(degenrate ellipse)为一个点

### 正圆
#### 圆心位于原点的
- 正圆解析式
    > $x_1^2 + x_2^2=r^2$
- 矩阵表示
    > $x^t x=r^2$ 

    > x=$\begin{bmatrix}
    x_1 \\
    x_2  
    \end{bmatrix}$
- 三角函数解析式
    >$\begin{cases}
    y_1=r \sin t \\
    x_1=r \cos t
    \end{cases}$
#### 圆心位于(c1,c2)的解析式
- 解析式
    > $(x_1-c_1)^2 + (x_2-c_2)^2 = r^2$
- 矩阵表示
    > $(x-c)^t (x-c)=r^2$ 

    > x=$\begin{bmatrix}
    x_1\\
    x_2  
    \end{bmatrix}$

    > c=$\begin{bmatrix}
    c_1\\
    c_2  
    \end{bmatrix}$
- 三角函数解析式
    > $\begin{cases}
    x_1=c_1 + r \cos t \\
    x_2=c_2 + r \sin t 
    \end{cases}$
### 半圆
上半圆解析式:
>f($x_1$)=$\sqrt{r^2 - x_1^2}$

下半圆解析式:
>f($x_1$)=-$\sqrt{r^2 - x_1^2}$

### 椭圆
椭圆的长轴是指连接椭圆上两点所获得最长线段2a,反之短轴2b
#### 正椭圆
- 椭圆的解析式
    > $\frac{x_1^2}{a^2} + \frac{x_2^2}{b^2}$=1 (a>b>0)
- 矩阵解析式
    > $x^t$ $\begin{bmatrix}
        {\frac{1}{a^2}} & 0\\
        0 & {\frac{1}{b^2}}
    \end{bmatrix}$ $x$ =1
- 三角函数解析式
    > $\begin{cases}
                x_1 = a \cos t \\
                x_2 = b \sin t
                \end{cases}$
#### 焦点
椭圆的焦点(focus)位于长轴上,该椭圆上任意一点P到两个焦点F1和F2的距离等于2a
>$\left\vert PF1 \right\vert$ + $\left\vert PF2 \right\vert$=2a
#### 中心移动($c_1$,$c_2$)
椭圆的解析式
> $\frac{(x_1-c_1)^2}{a^2}$ + $\frac{(x_2-c_2)^2}{b^2}$=1 (a>b>0)
#### 旋转椭圆
逆时针旋转t后得到的解析式
> $\frac{(x_1 \cos t + x_2 \sin t)^2}{a^2}$ +$\frac{(x_1 \sin t - x_2 \cos t)^2}{b^2}$ =1 

顺时针旋转t后得到的解析式
> $\frac{(x_1 \cos t - x_2 \sin t)^2}{a^2}$ +$\frac{(x_1 \sin t + x_2 \cos t)^2}{b^2}$ =1

### 抛物线
开口向上的抛物线解析式
> $4p x_2 = x_1^2$ (p>0)
这条抛物线的原点位于(0,0),焦点(focus)位于(0,p),准线(directix)位于y=-p,当p小于0开口向下.平面上,抛物线每一个点与焦点之间距离等于点和准线之前点距离.利用着一规律可以推导抛物线的解析式.

抛物线的平移

顶点位于(k,h)的解析式
>$ 4p( x_2 - k) = ( x_1 - h )^2$

### 双曲线
#### 解析式
> $\frac{x_1^2}{a^2} - \frac{x_2^2}{b^2}$ =1

焦点c可以通过 $c^2 = a^2 + b^2$ 计算出来

双曲线上任意一点P到两个焦点F1和F2的差值为2a

$\left\vert PF1 \right\vert$ - $\left\vert PF2 \right\vert$= $\pm 2a$

#### 渐近线,切线

> $x_2$ = $\pm \frac{b}{a} x_1$

渐近线的曲线在(- $\infty$,- $\frac{b}{a}$ ) ( $\frac{b}{a}$ , + $\infty$ )

#### 平移
将双曲线中心平移至(h,k)

> $\frac{(x_1-h)^2}{a^2} - \frac{(x_2-k)^2}{b^2}$ =1

```python
import matplotlib.pyplot as plt
from sympy import plot_implicit,symbols,Eq
x1,x2=symbols('x1,x2')

# curve 曲线
def plot_curve(Eq_expr):
    h_plot=plot_implicit(Eq_expr,(x1,-2.5,2.5),(x2,-2.5,2.5),xlabel='$\it{x_1}$',ylabel='$\it{x_2}$')

# 清除所有
plt.close('all')

# circle
Eq_sym=Eq(x1**2 + x2**2,1) 
plot_curve(Eq_sym)

# ellipse
Eq_sym=Eq(x1**2/4+x2**2,1)
plot_curve(Eq_sym)

# rotated counter 
Eq_sym=Eq(5*x1**2/8 + 3*x1*x2/4+5*x2**2/8,1)
plot_curve(Eq_sym)

import numpy
a=numpy.sin(numpy.deg2rad(135))
b=numpy.cos(numpy.deg2rad(135))

Eq_sym=Eq((x1*a+x2*b)**2/4+(x1*a-x2*b)**2,1)
plot_curve(Eq_sym)

Eq_sym=Eq(4*x1,x2**2)
plot_curve(Eq_sym)

Eq_sym=Eq(x1**2/2-x2**2/1,1)
plot_curve(Eq_sym)
```
